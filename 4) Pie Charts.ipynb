```python
import matplotlib.pyplot as plt
```

 # Popularity of different sports

```python
plt.style.use('fivethirtyeight')
scores=[70,100,80,50]
names=['Cricket','Football','Badminton','Hokey']
colors=['#F322B4','#1CE4D6','#EBF92E','#FC4A14']
explode=[0,0,0,0.1]
plt.pie(scores,labels=names,colors=colors, autopct='%0.2f%%', wedgeprops={'edgecolor': 'black'} ,explode=explode)  #edgecolor=black will add a blackcolor lines between two color.

plt.title('Popularity of different sports')
plt.tight_layout()
plt.legend(loc='upper left', fontsize=6)
plt.show()
```

   # Popularity of different programming languages

```python
plt.style.use('fivethirtyeight')
points = [59219, 55466, 47544, 36443, 35917]
languages=['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
colors=['#F322B4','#1CE4D6','#EBF92E','#FC4A14', '#CCFFCC']
explode=[0,0,0,0.1,0]
plt.pie(points,labels=languages,colors=colors, shadow=True,
        startangle=80, autopct='%1.1f%%',
        wedgeprops={'edgecolor': 'black',
                   'linewidth': 2} ,explode=explode)  #edgecolor=black will add a blackcolor lines between two color.


plt.title('Popularity of different programming languages')
plt.tight_layout()
plt.legend(loc= 'lower right', fontsize=7)
plt.show()
```

   # Area of continents

```python


# the slices are ordered and 
# plotted counter-clockwise: 
all_continents = ['South America','Australia', 'Africa','Antarctica','Asia', 'Europe', 'North America'] 

area = [10,15,10,5,25, 20, 15] 
explode = (0, 0.1, 0.1, 0,0.1,0.1,0) 

plt.pie(area, explode = explode, labels = all_continents, autopct = '%0.2f%%',startangle = 0, 
                     wedgeprops = {
                         "edgecolor" : "black", 
                         "linewidth" : 2, 
                         "antialiased": True}) 


plt.legend(loc= 'lower left', fontsize=7)
plt.axis('equal')   # Equal aspect ratio ensures # that pie is drawn as a circle. 

plt.show() 
```

   # GDP of five countries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('GDP.csv', encoding= 'unicode_escape')
print(df)
```

```python
x1=df.iloc[0:5,2]
x2=df.iloc[5:10,2]
x3=df.iloc[10:15,2]
print(x1)
print(x2)
print(x3)
```

```python
colors=['#F322B4','#1CE4D6','#EBF92E','#FC4A14', '#CCFFCC']
plt.figure(figsize=(8,6)) #Setting figure size
plt.pie(x1, labels=df.iloc[0:5,0], startangle=90,
       colors=colors, pctdistance=0.9, autopct='%1.1f%%, \n (2022)',
        radius=0.5, labeldistance=1.05, textprops={'fontweight':'bold', 'fontsize':8},
       wedgeprops={'linewidth':3, 'edgecolor': 'white'})

plt.pie(x2, startangle=90,
       colors=colors, pctdistance=0.85, autopct='%1.1f%%, \n (2021)',
        radius=0.4, textprops={'fontweight':'bold', 'fontsize':8},
       wedgeprops={'linewidth':3, 'edgecolor': 'white'})

plt.pie(x3, startangle=90,
       colors=colors, pctdistance=0.80, autopct='%1.1f%%, \n (2020)',
        radius=0.3, textprops={'fontweight':'bold', 'fontsize':6},
       wedgeprops={'linewidth':3, 'edgecolor': 'white'})


centre_circle=plt.Circle((0,0),0.20, fc='white')
plt.title('GDP of five countries')
fig=plt.gcf()
fig.gca().add_artist(centre_circle)
plt.axis('equal')
plt.legend(loc= 'upper left', fontsize=10)
plt.tight_layout()
plt.show()
```

# Donut chart

```python
plt.style.use('fivethirtyeight')
scores=[70,100,80,50]
names=['Cricket','Football','Badminton','Hokey']
colors=['#F322B4','#1CE4D6','#EBF92E','#FC4A14']
explode=[0,0,0,0.1]
plt.pie(scores,labels=names,colors=colors, autopct='%0.2f%%', wedgeprops={'edgecolor': 'black'} ,explode=explode)  #edgecolor=black will add a blackcolor lines between two color.

plt.title('Popularity of different sports')
plt.tight_layout()
plt.legend(loc='upper left', fontsize=6)


#Donut chart 
circle= plt.Circle(xy=(0,0), radius=.75, facecolor='white')
plt.gca().add_artist(circle)
plt.show()
```
