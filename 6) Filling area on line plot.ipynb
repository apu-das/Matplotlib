```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('data (1).csv')
ages=data['Age']
dev_salaries=data['All_Devs']
py_salaries=data['Python']

plt.plot(ages,dev_salaries,color='#F322B4', linestyle='--', label='All_Devs')
plt.plot(ages,py_salaries,color='#1CE4D6', label='Python')


median= 60000
plt.fill_between(ages, py_salaries, median,
                 where=(py_salaries>median), interpolate= True, alpha=0.50) 

plt.fill_between(ages, py_salaries, median,
                 wherAe=(py_salaries<=median), color='#1CE4D6' , interpolate= True, alpha=0.50)

plt.legend()
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Salaries Comparison among developers')
plt.tight_layout()
plt.show()
```

```python
data=pd.read_csv('data (1).csv')
ages=data['Age']
dev_salaries=data['All_Devs']
py_salaries=data['Python']

plt.plot(ages,dev_salaries,color='#F322B4', linestyle='--', label='All_Devs')
plt.plot(ages,py_salaries,color='#FC4A14', linestyle='dashed',label='Python')



plt.fill_between(ages, py_salaries,dev_salaries,
                 where=(py_salaries>dev_salaries), interpolate= True, alpha=0.50, label='Above Avg') 

plt.fill_between(ages, py_salaries,dev_salaries,
                 where=(py_salaries<=dev_salaries), color='#1CE4D6' , interpolate= True, alpha=0.50, label='Below Avg')

plt.legend()
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Salaries Comparison among developers')
plt.tight_layout()
plt.show()
```
