```python
import csv
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
```

```python
#Step-1: Count the users of each language

with open('data.csv') as csv_file:
    csv_reader=csv.DictReader(csv_file)
    
    language_counter=Counter()
    
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))
print(language_counter)
```

```python
#working with 15 most used languages

with open('data.csv') as csv_file:
    csv_reader=csv.DictReader(csv_file)
    
    language_counter=Counter()
    
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))
language_counter.most_common(15)
```

```python
#Step-2: Making a list of 15 most popular languages 

plt.style.use('fivethirtyeight')
with open('data.csv') as csv_file:
    csv_reader=csv.DictReader(csv_file)
    
    language_counter=Counter()
    
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))
        
languages=[]
popularity=[]      

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
print(languages)
print(popularity)
```

```python
#plotting bar
plt.style.use('fivethirtyeight')
with open('data.csv') as csv_file:
    csv_reader=csv.DictReader(csv_file)
    
    language_counter=Counter()
    
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))
        
languages=[]
popularity=[]      

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
    
    
languages.reverse()  # To plot most used languages in the top
popularity.reverse()

plt.barh(languages,popularity)
plt.title('Most popular programming language')
plt.xlabel('Number of user')
# plt.ylabel('languages')
plt.tight_layout()
plt.show()
```

   # Analyzing data with Pandas

```python
import csv
import numpy as np
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
```

```python
#plotting bar
plt.style.use('fivethirtyeight')

df=pd.read_csv('data.csv')
ids=df['Responder_id']
lang_responses= df['LanguagesWorkedWith']

language_counter=Counter()
    
for response in lang_responses:
        language_counter.update(response.split(';'))
        
languages=[]
popularity=[]      

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
    
    
languages.reverse()  # To plot most used languages in the top
popularity.reverse()

plt.barh(languages,popularity)
plt.title('Most popular programming language')
plt.xlabel('Number of user')
# plt.ylabel('languages')
plt.tight_layout()
plt.show()
```
